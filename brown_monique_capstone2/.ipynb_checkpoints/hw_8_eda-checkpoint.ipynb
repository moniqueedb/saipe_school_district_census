```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy 
import seaborn as sns

%matplotlib notebook
```

```python
%matplotlib inline
```

```python
hospital_ins = pd.read_csv(r'hospital-charges.csv', dtype= {'procedure': str, 'provider_id':int, 'provider_name': str, 'provider_address':str, 'provider_city': str, 'provider_state': str, 'zip_code': int, 'total_discharges': int, 'avg_covered_charges': int, 'avg_total_payments': int, 'avg_medicare_payments': int})
hospital_ins.head()
```

```python
## shape - how many rows*columns ##

hospital_ins.shape
```

```python
## what type of datatypes are in the dataset? ##

hospital_ins.dtypes
```

## Analysis

```python
hospital_ins.columns
```

```python
hospital_ins.plot()
```

```python
## distribution plot for avg_total_payments ##

sns.displot(hospital_ins['avg_total_pymnts'],bins=20)
```

```python
## distribution of zip codes of providers ##

sns.displot(hospital_ins['zip_code'], bins=30)
```

```python
## distribution plot of average amt of covered hospital charges ##

sns.displot(hospital_ins['avg_covered_charges'])
```

```python
## joint plot taking avg total payments made per zip code ##

sns.jointplot(data=hospital_ins,x='zip_code',y='avg_total_pymnts',kind='hex')
```

```python
sns.jointplot(data=hospital_ins,x='zip_code',y='avg_total_pymnts')
```

```python
sns.pairplot(hospital_ins)
```

```python
sns.lmplot(x='avg_total_pymnts',y='avg_medicare_pymnts',data=hospital_ins)
```
